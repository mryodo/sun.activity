# Положительные $\Delta d$

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import interpolate
import time

from math import *
```

```python
W=2*np.pi
dw=0.1
T=(2*np.pi)/W
n=15
L=n*T

N_grid=6000
t=np.linspace(0,L,N_grid)
h=t[1]-t[0]

p=int(round(T/h))
d=0.25
dd=0.25
```

```python
st=0
while (t[st]-T/2<0):
    st=st+1
fn=N_grid-1    
while (t[fn]+T/2>L):
    fn=fn-1
print(st, fn)
```

```python
jump_k=np.fromfile(file='/home/mryodo/gdrive/sun.activity/num_res/PT1806_15_6000_k.out', sep='\n')
jump_kC=np.fromfile(file='/home/mryodo/gdrive/sun.activity/num_res/PT1806_15_6000_kC.out', sep='\n')
jump_kR=np.fromfile(file='/home/mryodo/gdrive/sun.activity/num_res/PT1806_15_6000_kR.out', sep='\n')
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plt.figure(figsize=(7,7))
plt.grid('on')
plt.plot(1/p*np.array(range(p)), jump_k[:p])
#plt.plot(1/p*np.array(range(2*p)), jump_kC[:p])
plt.xlabel(r'{ \tau }/{ T }', fontsize=20)
plt.legend(['jumpK', 'jumpKC'], fontsize=15)
plt.ylabel(r'jumpK', fontsize=20)
plt.show()
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plt.figure(figsize=(7,7))
plt.grid('on')
plt.plot(1/p*np.array(range(p)), jump_kR[:p])
#plt.plot(1/p*np.array(range(2*p)), jump_kC[:p])
plt.xlabel(r'{ \tau }/{ T }', fontsize=20)
plt.legend(['jumpKR', 'jumpKC'], fontsize=15)
plt.ylabel(r'jumpKR', fontsize=20)
plt.show()
```

```python
sk=[]
for s in range(N_grid-2*p):
    k0=np.array([d]*N_grid)
    k0[2*p:2*p+s]=d+dd
    sk.append(np.std(k0))

jump_kR0=np.divide(jump_kR, sk)

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plt.figure(figsize=(7,7))
plt.grid('on')
plt.plot(1/p*np.array(range(p)), jump_kR0[:p])
plt.xlabel(r'{ \tau }/{ T }', fontsize=20)
plt.ylabel(r'jumpKR0', fontsize=20)
plt.show()
```
