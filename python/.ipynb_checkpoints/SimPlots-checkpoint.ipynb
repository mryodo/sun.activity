```r
if (!require("pacman")) install.packages("pacman")
pacman::p_load(ggplot2,reshape2,gridExtra,GenomeInfoDb,rtracklayer,dplyr)

PopulationDynamics10.4.2.01 <- read.csv("~/PopulationDynamics10.4.2.01.txt")
    
ggplot(data=PopulationDynamics10.4.2.01%>%
            filter(rep<5)%>%
            mutate(rep=as.factor(rep)),
        aes(x=time,y=count,color=rep))+
    geom_line()+
    labs(title="d=0.2 b=0.4 d'=0.01 L=10 sigma_d=sigma_w=0.04")

    
ggplot(data=PopulationDynamics10.4.2.01%>%
        filter(rep<30)%>%
        mutate(rep=as.factor(rep)),
        aes(x=time,y=count,color=rep))+geom_line()+
    labs(title="d=0.2 b=0.4 d'=0.01 L=10 sigma_d=sigma_w=0.04")
```
