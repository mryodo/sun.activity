```python
import scipy.integrate as integrate
import matplotlib.pyplot as plt
import numpy as np
from scipy import interpolate

from math import *
```

```python
N_grid=1000
L=10

W=np.pi/2
dw=0.1
```

```python
t=np.linspace(0, L, N_grid)
h=t[2]-t[1]
```

```python
k0=np.array([0.25]*N_grid)
k0[300:500]=0.5

plt.figure()
plt.plot(t, k0)
plt.show()
```

```python
k0_f=interpolate.interp1d(t, k0, bounds_error=False, fill_value="extrapolate")
```

```python
def deriv_z(z, t2, dw, k0_f):
    return 2*dw-sin(z)*k0_f(t2)
```

```python
init=0.2

theta = integrate.odeint(deriv_z, init, t, (dw, k0_f,))
theta=theta.ravel()
```

```python
x0=np.sin(W*t)
y0=np.sin(W*t+theta)
```

```python
plt.figure()
plt.plot(t, x0)
plt.plot(t, y0)
plt.show()
```

```python
T=2*np.pi/W
th=int(round(T/(2*h)))

C0=np.array([0]*N_grid)
x_с=np.append(x0, np.append(x0, x0))
y_с=np.append(y0, np.append(y0, y0))
for i in range(N_grid):
    a1=x_с[i-th+N_grid:i+th+N_grid]
    a2=y_с[i-th+N_grid:i+th+N_grid]
    C0[i]=np.corrcoef(a1, a2)[0][1]
    
phi0=np.arccos(C0)
C0
```

```python
a1=x_с[-th+N_grid:th+N_grid]
a2=y_с[-th+N_grid:th+N_grid]
test=np.corrcoef(a1, a2)[0][1]
bb[0]=test
bb[0]
```

```python
plt.figure()
plt.plot(t, theta)
plt.plot(t, phi0)
plt.show()
```

```python
dphi_hat=np.diff(phi0)/h
dphi_hat=np.append(dphi_hat[1], dphi_hat)
k_hat=np.divide(np.array([2*dw]*N_grid)-dphi_hat, np.sin(phi0))

for i in range(N_grid):
    if (abs(k_hat[i])>1):
        k_hat[i]=nan
        
plt.figure()
plt.grid(True)
plt.plot(t, k0)
plt.plot(t, k_hat)
plt.show()
```
